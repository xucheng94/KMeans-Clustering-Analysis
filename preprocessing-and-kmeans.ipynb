```python
#creat a mini dataset used for kmeans cluster, and make sure it does not influence the original data

temporary_data=data.copy()


temporary_data.head()
```

```python
# transfer some columns to numeric 



temporary_data['Chronic Medical Conditions'].replace({"Yes":1,"No":0},inplace=True)

temporary_data['Smoking Status'].replace({"Non-smoker":0,"Former":1,"Current":2},inplace=True)

temporary_data['Family History of Depression'].replace({"Yes":1,"No":0},inplace=True)

temporary_data['History of Substance Abuse'].replace({"Yes":1,"No":0},inplace=True)
```

```python
temporary_data['Age']=pd.qcut(data['Age'], q=4, labels=['0','1','2','3'])
```

```python
#sample all the features we need, and put them into a dataset

test_data=temporary_data[['Age','Chronic Medical Conditions','Smoking Status','Family History of Depression','History of Substance Abuse']]

test_data.head()
```

```python
#draw SEE graph so that we can find appropriate value for number of clusters

import matplotlib.pyplot as plt
from sklearn import cluster
%matplotlib inline
numClusters=[]
for num in range (1,9):
    numClusters.append(num)
SSE = []
for k in numClusters:
    k_means = cluster.KMeans(n_clusters=k)
    k_means.fit(test_data)
    SSE.append(k_means.inertia_)

plt.plot(numClusters, SSE)
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
```

From the SSE graph ,according to elbow therom(the point of inflection
on the curve,)we can get the best value of  is 3k

```python
#use the silhouette analysis to show results
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
from sklearn.utils import resample
from sklearn.decomposition import PCA
sampled_X = resample(test_data, n_samples=20000, random_state=42)
X = sampled_X

range_n_clusters = [2,3,4,5]
```

```python
for n_clusters in range_n_clusters:
    # create subplot
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)
    
    # set graph of Silhouette range
    ax1.set_xlim([-0.1, 1])
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # kmeans clustering
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # caculate Silhouette score
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"For n_clusters = {n_clusters}, the average silhouette_score is : {silhouette_avg}")
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # get i value from Silhouette and sort them in order
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )
        
        # label every clusters
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10  # increase space between current cluster and next cluster

    # set Silhouette's titles and label for x,y axis
    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # draw Silhouette average score line
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")
    ax1.set_yticks([])
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(
        X.iloc[:, 0], X.iloc[:, 1], marker=".", s=30, lw=0, alpha=0.7, c=colors, edgecolor="k"
    )

    # draw cluster center
    centers = clusterer.cluster_centers_
    ax2.scatter(
        centers[:, 0],
        centers[:, 1],
        marker="o",
        c="white",
        alpha=1,
        s=200,
        edgecolor="k",
    )
    
    # label cluster center
    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker=f"${i}$", alpha=1, s=50, edgecolor="k")

    # set title and axis for second axis
    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature 1")  
    ax2.set_ylabel("Feature 2")  

    # title
    plt.suptitle(
        f"Silhouette analysis for KMeans clustering on sample data with n_clusters = {n_clusters}",
        fontsize=14,
        fontweight="bold",
    )

plt.show()
```
